```r
download.file("http://www.openintro.org/stat/data/bdims.RData", destfile = "bdims.RData")
load("bdims.RData")
```

```r
head(bdims)
```

```r
mdims <- subset(bdims, sex == 1)
fdims <- subset(bdims, sex == 0)
```

## Exercise 1

```r
par(mfrow = c (1,2))
hist(mdims$hgt)
hist(fdims$hgt)
```

```r
fhgtmean <- mean(fdims$hgt)
fhgtsd   <- sd(fdims$hgt)
```

## Exercise 2

```r
hist(fdims$hgt, probability = TRUE, breaks = 20)
x <- 140:190
y <- dnorm(x = x, mean = fhgtmean, sd = fhgtsd)
lines(x = x, y = y, col = "blue")
```

It appears to be normally distributed

## Exercise 3

```r
qqnorm(fdims$hgt)
qqline(fdims$hgt)
```

```r
sim_norm <- rnorm(n = length(fdims$hgt), mean = fhgtmean, sd = fhgtsd)
qqnorm(sim_norm)
qqline(sim_norm)
```

The sim_norm follows better the line.

```r
qqnormsim <- function (dat) 
{
    par(mfrow = c(3, 3))
    qqnorm(dat, main = "Normal QQ Plot (Data)")
    qqline(dat)
    for (i in 1:8) {
        simnorm <- rnorm(n = length(dat), mean = mean(dat), sd = sd(dat))
        qqnorm(simnorm, main = "Normal QQ Plot (Sim)")
        qqline(simnorm)
    }
    par(mfrow = c(1, 1))
}
```

```r
qqnormsim(fdims$hgt)
```

```r
qqnormsim(fdims$wgt)
```

```r
qqnorm(fdims$wgt)
qqline(fdims$wgt)
```

The female weight distribution, unlike the height distribution, does not appear to be normally distributed relative to the simulated data. The female weight seems to deviate wildly outside of the median
